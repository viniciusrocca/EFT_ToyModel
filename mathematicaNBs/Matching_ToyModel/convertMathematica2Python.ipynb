```python
from sympy.parsing.mathematica import parse_mathematica
from sympy import var
mSDM = var('mSDM')
mPsiT = var('mPsiT')
g1,g2,g3,yDM,lamHs,lamS = var('g1,g2,g3,yDM,lamHs,lamS')
```

```python
with open('coeffs_py.txt','r') as f:
    lines = f.readlines()

exprDict = {}
for l in lines:
    entries = l.split(";")
    coeff = entries[0]
    indices = [s.replace('_','') for s in entries[1:-2]]
    expr = entries[-1]

    python_code = parse_mathematica(expr)

    exprDict[coeff] = {'indices' : indices, 'expr' : str(python_code)}
```

```python
import json
with open('../Plots/coeffs_python.json','w') as f:
    json.dump(exprDict,f)
```
