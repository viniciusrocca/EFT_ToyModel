```python
import numpy as np
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings("ignore", message="numpy.dtype size changed")
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

pd.option_context('display.max_columns', -1)

pd.options.mode.chained_assignment = None #Disable copy warnings
# plt.style.use('fivethirtyeight') #Set style
# mpl.rcParams.update({'figure.figsize' : (15,10)})  #Set general plotting options
plt.rcParams['figure.max_open_warning'] = 50
plt.rcParams.update({
    "text.usetex": True,
    "font.family": "sans-serif",
    "font.sans-serif": ["Helvetica"]})

plt.rcParams.update({"savefig.dpi" : 300}) #Figure resolution


#Define plotting style:
sns.set() #Set style
sns.set_style('ticks',{'font.family':'Times New Roman', 'font.serif':'Times New Roman'})
sns.set_context('paper', font_scale=1.8)
cm = plt.colormaps['RdYlBu']
```

```python
def L(s, mT_squared):
    """Calculates the function L(s, m_T^2)."""
    # Cast inputs to complex type to enable complex math
    s = np.array(s, dtype=complex)
    mT_squared = complex(mT_squared)
    
    sqrt_term = np.sqrt(s * (s - 4 * mT_squared))
    numerator = sqrt_term + 2 * mT_squared - s
    denominator = 2 * mT_squared
    
    return np.log(numerator / denominator)
```

```python
def F(s, mT_squared):
    """Calculates the function F(s, m_T^2)."""
    # Cast inputs to complex type
    s = np.array(s, dtype=complex)
    mT_squared = complex(mT_squared)
    
    L_val = L(s, mT_squared)
    
    term1 =  mT_squared * L_val**2
    term2 = np.sqrt(s * (s - 4 * mT_squared)) * L_val
    term3 = 3 * s
    
    return (1 / s**2) * (term1 + term2 + term3)
```

```python
#Seting the mass parameter for the colored scalar
m_T = 100.0
mT_squared = m_T**2

#Setting the range of the plot
s_over_mT_squared = np.linspace(0.1, 100, 1000)

#Computing the s values to use in F
s_values = s_over_mT_squared * mT_squared

#Computing F
F_values_complex = F(s_values, mT_squared)


#Spliting the real, complex and the abs value of F
F_real = np.real(F_values_complex)
F_imag = np.imag(F_values_complex)
F_abs = np.abs(F_values_complex)

#Setting the plot
plt.style.use('seaborn-v0_8-whitegrid')
fig, ax = plt.subplots(figsize=(10, 6))

# Ploting the results
ax.plot(s_over_mT_squared, F_real, label=r'Re($\mathcal{F}$)')
ax.plot(s_over_mT_squared, F_imag, label=r'Im($\mathcal{F}$)')
ax.plot(s_over_mT_squared, F_abs, label=r'$|\mathcal{F}|$')
ax.axvline(x=4.0, color='gray', linestyle='--', alpha=0.7, label=r'$s=4m_T^2$')

ax.set_xlabel(r'$s/m_T^2$', fontsize=14)
ax.set_ylabel('Arbitrary Units', fontsize=14)
ax.set_xlim(0, 100)
#ax.set_yscale('log')
ax.legend(fontsize=14)
plt.show()
```

```python
# 1. Define the mass values and corresponding plot styles
masses_to_compare = [0.175, 0.200]
colors = ['C0', 'C3'] # Blue for the first mass, Red for the second
linestyles = ['-', '--', ':'] # Solid for Re(F), Dashed for Im(F), Dotted for |F|

# 2. Setting the plot
plt.style.use('seaborn-v0_8-whitegrid')
fig, ax = plt.subplots(figsize=(12, 7))

# 3. Define a common, absolute energy range for the x-axis
sqrt_s = np.linspace(0.1, 1, 1000) # Plotting from 1 to 4000 GeV
s_values = sqrt_s**2

# 4. Loop through each mass to calculate and plot its curves
for m_T, color in zip(masses_to_compare, colors):
    mT_squared = m_T**2
    
    # Computing F using the common s_values
    F_values_complex = F(s_values, mT_squared)

    # Splitting the real, imaginary, and the abs value of F
    F_real = np.real(F_values_complex)
    F_imag = np.imag(F_values_complex)
    F_abs = np.abs(F_values_complex)
    
    # Ploting the results for the current mass
    label_prefix = f'$m_T = {m_T}$ GeV'
    ax.plot(sqrt_s, F_real, color=color, linestyle=linestyles[0], label=f'{label_prefix}: Re($\\mathcal{{F}}$)')
    ax.plot(sqrt_s, F_imag, color=color, linestyle=linestyles[1], label=f'{label_prefix}: Im($\\mathcal{{F}}$)')
    ax.plot(sqrt_s, F_abs, color=color, linestyle=linestyles[2], label=f'{label_prefix}: $|\\mathcal{{F}}|$')

    # Add a vertical threshold line specific to this mass
    threshold = np.sqrt(4.0 * mT_squared)
    ax.axvline(x=threshold, color=color, linestyle='--', alpha=0.5)


# 5. Finalize the plot
ax.set_xlabel(r'$\sqrt{s}$ [TeV]', fontsize=14)
ax.set_ylabel('Arbitrary Units', fontsize=14)
ax.set_xlim(left = 0, right=1)
#ax.set_ylim(0, 0.0000025) # Adjust y-limit to better see the features
ax.legend(fontsize=12, ncol=2)
plt.tight_layout()
plt.savefig('loop_function_F_full_comparison_dimensionful.png')
plt.show()
```
